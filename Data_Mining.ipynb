```python
import joblib

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
```

```python
# Load the saved model
svm_body_model = joblib.load('data/svm_body_model.pkl')
```

```python
# Run headless (no browser window)
chrome_options = Options()
chrome_options.add_argument("--headless")

# Path to your chromedriver
service = Service('path/to/chromedriver')

driver = webdriver.Chrome(service=service, options=chrome_options)
```

```python
url = "https://www.example-news-site.com/some-article"
driver.get(url)

# Grab article text (change the selector to match the website)
article_elements = driver.find_elements(By.TAG_NAME, "p")  # usually paragraphs
article_text = " ".join([elem.text for elem in article_elements])
```
